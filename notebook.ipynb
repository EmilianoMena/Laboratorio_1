## Introducción
En este proyecto se busca comparar dos tipos de inversión (Pasiva y Activa) y poder ver cuales son las principales diferencias entre ambas estrategias.

### Inversión Pasiva
Por un lado tenemos la Inversión Pasiva que se puede considerar como la opción mas conservadora. De lo que se trata es de seguir un indice en especifico para obtener un rendimiento de mercado. Esta estrategia es apoyada por diversas figuras del mundo de las finanzas como:

-Eugene Fama: Que en su teoría de Mercados Eficientes menciona que tratar de superar los rendimientos que ofrece el mercado es más cuestión de suerte que de habilidad. Esto lo afirma al decir que los precios de los activos son en si toda la información de dichos activos y aparte el Mercado al ser eficiente pues no se le puede ganar.

-William Sharpe: El cual dice que un fondo indexado es el que ofrece el mayor rendimiento a menor riesgo.

-Harry Markowitz: Menciona que se debe considerar no solo el rendimiento al hora de invertir si no que tambien tomar en cuenta el riesgo que involucra dicha inversión. Y para poder disminuir este riesgo se debe diversificar la inversión. 

Basicamente la base de esta estrategia de inversión es que no se puede superar los rendimientos del mercado, por ende la opción de indices que siguen el comportamiento del mercado es la mejor forma de invertir sabiendo que no vas a perder.

### Inversión Activa
Por otro lado esta la Inversión Activa que es una estrategia en la cual no te atienes a un indice para invertir, si no más bien que seleccionas los activos por cuenta propia. Esto te permite poder elegir las acciones que desees y no solo las de un mercado, esto a su vez te brinda la oportunidad de mitigar riesgos al evitar ciertos sectores que se incluyen en indices que son de riesgo. Y lo principal de esta estrategia es la busqueda de maximizar los rendimientos para poder superar los que ofrece el mercado.
